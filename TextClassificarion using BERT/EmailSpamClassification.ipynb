# **Creating a Embedding using BERT **

Creating a preprocess and encoder layer

```python
bert_preprocess = hub.KerasLayer(
    "https://kaggle.com/models/tensorflow/bert/TensorFlow2/en-uncased-preprocess/3")

bert_encoder = hub.KerasLayer(
    "https://tfhub.dev/tensorflow/bert_en_uncased_L-12_H-768_A-12/4",
    trainable=True)
```

```python
def get_sentence_embeding(sentences):
    preprocessed_text = bert_preprocess(sentences)
    return bert_encoder(preprocessed_text)['pooled_output']
```

```python
get_sentence_embeding([
    "Hi ,How are you babe",
    "Babu, are you busy today"]
)
```

```python
example= get_sentence_embeding([
    "banana",
    "pineapple",
    "mango",
    "dhoni",
    "virat",
    "sachin"])
```

Comparision using Cosine-Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity

print(cosine_similarity([example[0]], [example[1]]))
print(cosine_similarity([example[0]], [example[2]]))
print(cosine_similarity([example[3]], [example[4]]))
print(cosine_similarity([example[5]], [example[4]]))
print(cosine_similarity([example[0]], [example[3]]))
```

```python
# Building a Functional Model
# two types of model - Seq and fuctional
# Seq model=adding layer seq by seq
# func model=create inp feed it into the hidden layer n times and create a model using both input and output layers


text_input = tf.keras.layers.Input(shape=(), dtype=tf.string, name='text')
preprocessed_text = bert_preprocess(text_input)  # Preprocess only once
outputs = bert_encoder(preprocessed_text)  # Use preprocessed text here

net = outputs['pooled_output']

l = tf.keras.layers.Dropout(0.1, name="dropout")(net)
l = tf.keras.layers.Dense(1, activation="sigmoid", name='output')(l)  # Changed activation to lowercase

model = tf.keras.Model(inputs=[text_input], outputs=[l])
```
